```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Create bounding box for cat eyes
Using cat coord in Kaggle cat dataset

```python
# Get filename for all .cat files in folder CAT_00
import glob
import os

# Read all the data file names
os.chdir(r'CAT_00')
myFiles = glob.glob('*.jpg')
```

```python
len(myFiles)
```

```python
# Test read first file
file = myFiles[0]
img = cv.imread(file)
# BGR to RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# Default color in matplotlib is RGB
plt.imshow(rgb)
plt.show()
```

```python
# Test read file_coord
file_coord = file+".cat"
f = open(file_coord, "r")
coord = f.read().split(" ")
eye_1 = [int(x) for x in coord[1:3]]
eye_2 = [int(x) for x in coord[3:5]]
nose = [int(x) for x in coord[5:7]]
ear_1 = [int(x) for x in coord[7:9]]
ear_2 = [int(x) for x in coord[9:11]]
ear_3 = [int(x) for x in coord[11:13]]
ear_4 = [int(x) for x in coord[13:15]]
ear_5 = [int(x) for x in coord[15:17]]
ear_6 = [int(x) for x in coord[17:19]]
```

```python
# Draw a rectangle
box_size = (eye_1[0] - eye_2[0])//4
cv.rectangle(rgb,(eye_1[0]+box_size,eye_1[1]-box_size),\
                 (eye_1[0]-box_size,eye_1[1]+box_size),\
                 (0,255,0), thickness=2);
cv.rectangle(rgb,(eye_2[0]+box_size,eye_2[1]-box_size),\
                 (eye_2[0]-box_size,eye_2[1]+box_size),\
                 (0,255,0), thickness=2);
plt.imshow(rgb)
plt.show();
```

```python
for file in myFiles[:5]:
    # Test read first file
    img = cv.imread(file)
    # BGR to RGB
    rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # Test read file_coord
    file_coord = file+".cat"
    f = open(file_coord, "r")
    coord = f.read().split(" ")
    eye_1 = [int(x) for x in coord[1:3]]
    eye_2 = [int(x) for x in coord[3:5]]

    # Draw a rectangle
    box_size = (eye_1[0] - eye_2[0])//4
    cv.rectangle(rgb,(eye_1[0]+box_size,eye_1[1]-box_size),\
                     (eye_1[0]-box_size,eye_1[1]+box_size),\
                     (0,255,0), thickness=2);
    cv.rectangle(rgb,(eye_2[0]+box_size,eye_2[1]-box_size),\
                     (eye_2[0]-box_size,eye_2[1]+box_size),\
                     (0,255,0), thickness=2);
    plt.imshow(rgb)
    plt.show();
```

### Convert to YOLO files
After testing with 5 photos, let's write the .txt files.

```python
for file in myFiles:
    # jpg file
#     print(file)
    img = cv.imread(file)
    
    # .cat file to read coord from
    file_coord = file+".cat"
#     print(file_coord) 
    f = open(file_coord, "r")
    coord = f.read().split(" ")
    eye_1 = [int(x) for x in coord[1:3]]
    eye_2 = [int(x) for x in coord[3:5]]
    
    # txt file to write in
#     print(file[:-4]+".txt")
    
    # Box size depends on eye distance
    box_size = (eye_1[0] - eye_2[0])//4
    # Normalize the coord with image size
    with open(os.path.join('/Users/katiehuang/Documents/OpenCV/YOLO',file[:-4]+".txt"), "w") as file1:
        file1.write("0 "+str((eye_1[0]+box_size)/img.shape[0]) +" "+str((eye_1[1]-box_size)/img.shape[1])\
           +" " +str((eye_2[0]+box_size)/img.shape[0]) +" "+str((eye_2[1]-box_size)/img.shape[1]))
```
